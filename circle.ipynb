```python
"""
Solving the system of equations
x^2 + y^2 = 45
x*y=18

"""

import matplotlib.pyplot as plt
import numpy as np
import math

# Create a circle
circle = plt.Circle((0, 0), math.sqrt(45), color='b', fill=False)

# Create a plot
fig, ax = plt.subplots()
ax.add_artist(circle)

# Set the aspect of the plot to be equal, so the circle looks like a circle
ax.set_aspect('equal')

# Set the limits of the plot
ax.set_xlim(-math.sqrt(45), math.sqrt(45))
ax.set_ylim(-math.sqrt(45), math.sqrt(45))

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

"""
Solving the system of equations
x^2 + y^2 = 45
x*y=18

"""

# Create a circle
circle = plt.Circle((0, 0), math.sqrt(45), color='b', fill=False,label='$x^2$ + $y^2$ = 45')

# Generate x values for the curve
x = np.linspace(-12, 12, 300)  # Generate 300 points between -10 and 10

# Calculate corresponding y values based on the equation x*y=18
y = 18 / x

# Create a plot
fig, ax = plt.subplots()
ax.add_artist(circle)
ax.plot(x, y, label='x*y=18')
ax.set_aspect('equal')
ax.set_xlim(-12, 12)
ax.set_ylim(-12, 12)
ax.locator_params(axis='y', nbins=20) #number of ticks for y-axis
ax.locator_params(axis='x', nbins=12) #number of ticks for x-axis
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$x^2$ + $y^2$ = 45 and x*y=18')
ax.legend(loc='upper right')

# Display the plot
plt.grid(True)
plt.axhline(0, color='red',linewidth=0.5)
plt.axvline(0, color='green',linewidth=0.5)
plt.show()
```
